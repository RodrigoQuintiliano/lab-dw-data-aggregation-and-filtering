This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
print(df.shape) #10910 rows, 26 Columns.
df.head()
```

1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
# lets clean all columns names to avoid errors.
df.columns = (
    df.columns
      .str.strip()           # remove leading/trailing spaces
      .str.lower()           # lowercase
      .str.replace(' ', '_') # spaces to underscores
)
```

```python
df.nunique() # We can see we have a total of 9134 Customers from the 10910 rows, the difference are nan values.
```

```python
df['response'].unique() 
```

- We have 3 values in 'response' column, 
  (Yes, No and NaN), this means the column needs no cleaning, there are no typing errors only nan's, 
   we will ignore NaN values for now.

```python
df['response'].isna().sum() 
```

- 631 nan values from the total data frame, this confirms we can ignore nan values.

```python
# New Data Frame:

new_df = df[
    (df['total_claim_amount'] < 1000) & 
    (df['response'] == "Yes")
]

new_df.head()
```

```python
print(new_df.nunique()) 
```

```python
new_df['response'].value_counts(dropna=False)
```

- After filtering, from the total 9134 customers, we only have 1248 Clients who said 'Yes'.
- But 'Response' column gives us 1399 'Yes' counts, which means some clients said Yes more then once.

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
yes_clients = df[df['response'] == 'Yes']
```

```python
mean_values = yes_clients.groupby(['policy_type', 'gender'])[
    ['monthly_premium_auto', 'customer_lifetime_value']
].mean()

yes_clients.head()
```

```python
# Comparing yes clients with total claim amount:

claims_df = (
    yes_clients.groupby(['policy_type', 'gender'])
          .agg(avg_total_claim_amount=('total_claim_amount', 'mean'))
          .reset_index()
          .sort_values('avg_total_claim_amount', ascending=True) # Sorting from the less claim amount to most. 
)
claims_df['avg_total_claim_amount'] = claims_df['avg_total_claim_amount'].round(2) # Only 2 decimals

claims_df
```

- Corporate Auto: Females have the bigger claim amount.
- Personal Auto: Males have the bigger claim amount.
- Special Auto: Females have the bigger claim amount.
- By gender in general, Males have a total of 1295,12 average claim amount and Females a total of 1339,99 claim amount, which makes males the most profitable and low risk.

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
clv_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median'])

clv_stats = clv_stats.round(2)
clv_stats

```

### Education level strongly influences CLV
    - Bachelor degree customers in both genders show the highest maximum CLV values, up to 73225 for females and 67907 for males.
    - High school or below male customers show the single highest CLV values, up to 83325, indicating extremely valuable individuals.
    - After College customers, Doctors are the ones with the lowest max CLV values, indicating these groups do not include high-value outliers.
## Overall conclusion:
### The most valuable segments (based on median CLV) appear to be:
    - High School or Below – Male
    - High School or Below – Female
    - College – Male

### Meanwhile, the least valuable segments (median-wise) are:
    - Doctor : Female
    - Doctor : Male
This suggests that lifetime value is not strictly determined by education level, lower-education groups can still produce highly profitable long-term customers.
